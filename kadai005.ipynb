```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)
axes[0,0].set_title('plot')

data2 = np.random.normal(10,5,1000)
axes[0,1].hist(data2, bins=20)
axes[0,1].set_title('Histogram')

x3 = np.random.randn(100)
y3 = np.random.randn(100)
axes[0,2].scatter(x3,y3)
axes[0,2].set_title('Scatter Plot')

x4 = np.linspace(-5,5,100)
y4 = x4**2

axes[1,1].plot(x4, y4)
axes[1,1].set_title('Quadratic Function')



plt.show()





# x = np.linspace(0, 10, 100)
# y = 2 + 2*np.sin(1 * x)
# axes[0,1].plot(y)

# x = np.random.normal(15,5,200)
# plt.hist(x,bins=10)
# ax




# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
# x = np.random .normal(15, 5, 2000)
# plt.xlim(-10, 40)
# # ヒストグラムをプロット
# plt.hist(x, bins=30)
# axes[0,2].plot(y)


```
